Question 9 :- Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, 
company name, CTC, and apply date.


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r= requests.get('https://internshala.com/fresher-jobs')
r
```

```python
page = 1
Title = []

while page !=8:
    
    url = f"https://internshala.com/fresher-jobs/page={page}"
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html,'lxml')
    for div in soup.find_all('div',class_= 'heading_4_5 profile'):
        Title.append(div.text)
    page = page+1

Title
```

```python
len(Title)
```

```python
page = 1

Company = []

while page !=8:
    
    url = f'https://internshala.com/fresher-jobs/page={page}'
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
    for i in soup.find_all('a', class_='link_display_like_text'):
        Company.append(i.text.replace('\n',''))
        
    page = page+1


Company
```

```python
len(Company)
```

```python
page = 1
data= []

while page !=8:
    
    url = f'https://internshala.com/fresher-jobs/page={page}'
    r = requests.get(url)
    soup= BeautifulSoup(r.content)
    
    for i in soup.find_all('div', class_='item_body'):
        data.append(i.text)
        
    page = page+1


data
```

```python
len(data)
```

```python
CTC = data[1:840:3]

CTC
```

```python
len(CTC)
```

```python
Apply_Date = data [2:840:3]
Apply_Date
```

```python
len(Apply_Date)
```
